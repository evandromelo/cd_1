Script que armazena as temperaturas e os horários em duas listas separadas. 

Em seguida, ele calcula a média das temperaturas e, por fim, combina as duas listas em um dicionário para facilitar a consulta. 

Os resultados são impressos no final.

Para usar seus próprios dados, basta substituir os valores nas listas "temperaturas_celsius" e "horarios".

```python
# -*- coding: utf-8 -*-

# 1. Armazena os dados de temperatura e horário em listas
temperaturas_celsius = [65.2, 66.1, 67.5, 68.0, 67.2, 66.8]
horarios = ["14:00", "14:15", "14:30", "14:45", "15:00", "15:15"]

# 2. Calcula a média das temperaturas
# A função sum() soma todos os itens da lista e len() conta quantos itens existem.
media_temperaturas = sum(temperaturas_celsius) / len(temperaturas_celsius)

# 3. Armazena os dados em um dicionário
# A função zip() combina as duas listas, criando pares de chave-valor.
dados_do_secador = dict(zip(horarios, temperaturas_celsius))

# 4. Imprime os resultados de forma organizada
print("--- Análise de Dados do Secador Experimental ---")
print("\nLista de temperaturas coletadas (°C):")
print(temperaturas_celsius)

print(f"\nA média das temperaturas é: {media_temperaturas:.2f}°C")

print("\nRegistro de temperaturas por horário:")
# Imprime o dicionário de forma mais legível
for horario, temperatura in dados_do_secador.items():
    print(f" - {horario}: {temperatura}°C")
```

```python
# -*- coding: utf-8 -*-

# Importa a biblioteca NumPy, comum para computação numérica
import numpy as np

# 1. Modela a matriz de temperaturas
# A estrutura é a mesma (lista de listas), que será convertida para um array NumPy
matriz_de_temperaturas = [
    [25.5, 26.1, 25.8, 26.0],
    [27.3, 27.5, 28.0, 27.8],
    [24.9, 25.0, 25.3, 25.1],
]

# 2. Converte a lista Python para um array NumPy
# NumPy oferece estruturas de dados e funções otimizadas para arrays
array_de_temperaturas = np.array(matriz_de_temperaturas)

# 3. Calcula a média usando a função np.mean()
# Este método é altamente otimizado para calcular a média de todos os elementos do array
media_geral = np.mean(array_de_temperaturas)

# 4. Imprime os resultados
print("--- Análise da Matriz de Temperaturas (com NumPy) ---")
print("\nMatriz de Temperaturas Coletadas (°C):")
# Imprime a matriz de forma mais legível
for linha in matriz_de_temperaturas:
    print(linha)

print(f"\nTemperatura Média Geral: {media_geral:.2f}°C")
print("\n(Cálculo realizado utilizando a biblioteca NumPy)")
```

Script que utiliza as bibliotecas Pandas e NumPy   
para simular dados de rendimento de óleo essencial   
em diferentes temperaturas de extração.

A simulação assume que o rendimento aumenta com a temperatura  
até um ponto ótimo e depois começa a diminuir,   
o que pode ocorrer devido à degradação de compostos.  
Foi adicionada também uma pequena variação aleatória  
para que os dados pareçam mais realistas,   
como se tivessem sido coletados em um experimento real.

**Analise estatistica DIC - dados simulados de um experimemnto**

**Como o Código Funciona:**
**Simulação dos Dados:** Usamos a biblioteca numpy para criar três conjuntos de dados que simulam o teor de canabinoides para cada temperatura. Eu ajustei as médias para que houvesse uma diferença entre os grupos, imitando um resultado experimental real onde a temperatura ótima seria 50°C.

**ANOVA (stats.f_oneway):** Este é o primeiro e mais importante teste. Ele avalia todos os grupos ao mesmo tempo e nos dá um único p-valor. Se este valor for baixo (geralmente < 0.05), significa que a temperatura de secagem teve um efeito no teor de canabinoides.

**Teste t (stats.ttest_ind):** Se a ANOVA confirmar que existe uma diferença, este segundo passo é executado. O código compara cada par de tratamentos (45°C vs. 50°C, 45°C vs. 55°C, e 50°C vs. 55°C) para identificar exatamente onde estão as diferenças significativas.

Ao executar este script, você terá uma análise estatística robusta e uma interpretação clara dos resultados simulados. Você pode alterar os valores de loc (média) e scale (desvio padrão) no início do código para simular diferentes cenários experimentais.

```python
# -*- coding: utf-8 -*-

# Importa as bibliotecas necessárias
import numpy as np
from scipy import stats

# --- 1. Simulação dos Dados do Experimento ---
# Delineamento Inteiramente Casualizado (DIC)
# Medida: Teor de canabinoides (%)
# Tratamentos: 3 temperaturas de secagem

# Parâmetros da simulação
np.random.seed(42) # Define uma semente para que os resultados aleatórios sejam sempre os mesmos
num_amostras_por_tratamento = 10 # Número de repetições (plantas) por tratamento

# Simulando os dados para cada tratamento
# Vamos supor que 50°C é uma temperatura ótima, 45°C é boa e 55°C já degrada um pouco.
# loc = média do teor de canabinoides; scale = desvio padrão
tratamento_45C = np.random.normal(loc=18.5, scale=1.5, size=num_amostras_por_tratamento)
tratamento_50C = np.random.normal(loc=20.5, scale=1.2, size=num_amostras_por_tratamento)
tratamento_55C = np.random.normal(loc=17.0, scale=1.8, size=num_amostras_por_tratamento)

# Imprime os dados simulados para conferência (VERSÃO CORRIGIDA)
print("--- Dados Simulados do Teor de Canabinoides (%) ---")
print(f"Tratamento 45°C: {np.round(tratamento_45C, 2)}")
print(f"Tratamento 50°C: {np.round(tratamento_50C, 2)}")
print(f"Tratamento 55°C: {np.round(tratamento_55C, 2)}")
print("-" * 50)


# --- 2. Análise Estatística ---

# Passo A: Análise de Variância (ANOVA)
# A ANOVA é o teste correto para comparar 3 ou mais grupos.
# Hipótese Nula (H0): As médias de todos os tratamentos são iguais.
# Hipótese Alternativa (H1): Pelo menos uma média é diferente.
print("\n--- Passo A: Análise de Variância (ANOVA) ---")
f_statistic, p_value_anova = stats.f_oneway(tratamento_45C, tratamento_50C, tratamento_55C)

print(f"Estatística F: {f_statistic:.4f}")
print(f"P-valor da ANOVA: {p_value_anova:.4f}")

# Interpretação da ANOVA
alpha = 0.05
if p_value_anova < alpha:
    print("\nResultado ANOVA: Rejeitamos a hipótese nula (p < 0.05).")
    print("Conclusão: Existe uma diferença estatisticamente significativa entre as médias dos tratamentos de secagem.")
    prosseguir_com_t_test = True
else:
    print("\nResultado ANOVA: Não rejeitamos a hipótese nula (p >= 0.05).")
    print("Conclusão: Não há evidência de diferença significativa entre as médias dos tratamentos.")
    prosseguir_com_t_test = False
print("-" * 50)


# Passo B: Testes t par-a-par (Post-Hoc)
# Este passo só é executado se a ANOVA for significativa.
if prosseguir_com_t_test:
    print("\n--- Passo B: Testes t para Comparações Múltiplas ---")
    print("Comparando os pares de tratamentos para ver onde estão as diferenças.\n")
    
    # Comparações
    comparacoes = [
        ("45°C", "50°C", tratamento_45C, tratamento_50C),
        ("45°C", "55°C", tratamento_45C, tratamento_55C),
        ("50°C", "55°C", tratamento_50C, tratamento_55C)
    ]

    for nome1, nome2, dados1, dados2 in comparacoes:
        # Teste t para amostras independentes
        t_stat, p_value_t = stats.ttest_ind(dados1, dados2)
        
        print(f"Comparação: {nome1} vs {nome2}")
        print(f"  - Estatística t: {t_stat:.4f}")
        print(f"  - P-valor do Teste t: {p_value_t:.4f}")
        
        # Interpretação do teste t
        if p_value_t < alpha:
            print(f"  - Conclusão: Diferença SIGNIFICATIVA encontrada entre {nome1} e {nome2}.")
        else:
            print(f"  - Conclusão: Diferença NÃO significativa entre {nome1} e {nome2}.")
        print()
else:
    print("\nComo a ANOVA não foi significativa, não se prossegue com os testes t par-a-par.")

```
